```python
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import MultiPolygon, Polygon

import colors
```

```python
# Load the provinces shapefile
provinces = gpd.read_file("GADM Italy/gadm41_ITA_2.shp")
```

```python
provinces
```

```python
provinces_one = gpd.read_file("GADM Italy/gadm41_ITA_3.shp")
provinces_one = provinces_one[provinces_one["NAME_1"] == "Emilia-Romagna"]
provinces_one[provinces_one["NAME_2"] == "Reggio Nell'Emilia"]
```

```python
# Filter for Emilia-Romagna region
emilia_romagna_provinces = provinces[provinces["NAME_1"] == "Emilia-Romagna"]
```

```python
# Re-project to a suitable projected CRS
emilia_romagna_provinces = emilia_romagna_provinces.to_crs(epsg=3857)

# Calculate centroids
centroids = emilia_romagna_provinces.geometry.centroid
```

```python
emilia_romagna_provinces
```

```python
# Set the general font properties
plt.rcParams["font.family"] = "General Sans Variable"
plt.rcParams["font.size"] = 12
plt.rcParams["figure.facecolor"] = colors.CREAM  # Default figure background color
plt.rcParams["axes.facecolor"] = colors.CREAM  # Default axes background color
plt.rcParams["text.color"] = colors.BLACK_INK  # Default text color
plt.rcParams["axes.labelcolor"] = colors.BLACK_INK  # Default axes labels color
plt.rcParams["xtick.color"] = colors.BLACK_INK  # Default x-tick color
plt.rcParams["ytick.color"] = colors.BLACK_INK  # Default y-tick color
plt.rcParams["axes.titlecolor"] = colors.BLACK_INK  # Default title color
```

```python
# Plot the map
fig, ax = plt.subplots(figsize=(20, 20))
emilia_romagna_provinces.plot(ax=ax, color=colors.BLUE, edgecolor=colors.BLACK_INK)

# Add labels
for x, y, label in zip(
    emilia_romagna_provinces.geometry.centroid.x,
    emilia_romagna_provinces.geometry.centroid.y,
    emilia_romagna_provinces["NAME_2"],
    strict=False,
):
    if label == "Reggio Nell'Emilia":
        ax.text(x, y, label, fontsize=10, ha="center")
    else:
        ax.text(x, y, label, fontsize=10)


plt.title("Provinces of Emilia-Romagna, Italy")
plt.axis("off")
plt.savefig("plots/Provinces of Emilia-Romagna, Italy", bbox_inches="tight")
plt.show()
```

```python
# Define Romagna provinces
romagna_provinces = ["Ferrara", "Forli' - Cesena", "Ravenna", "Rimini"]

# Re-project to a suitable projected CRS
emilia_romagna_provinces = emilia_romagna_provinces.to_crs(epsg=3857)

# Separate Romagna provinces
romagna = emilia_romagna_provinces[
    emilia_romagna_provinces["NAME_2"].isin(romagna_provinces)
]
non_romagna = emilia_romagna_provinces[
    ~emilia_romagna_provinces["NAME_2"].isin(romagna_provinces)
]
```

```python
# Plot the map
fig, ax = plt.subplots(figsize=(20, 20))
non_romagna.plot(ax=ax, color=colors.BLUE, edgecolor=colors.BLACK_INK)
romagna.boundary.plot(ax=ax, color=colors.BLACK_INK)

# Add diagonal hatch pattern for Romagna provinces
for geom in romagna.geometry:
    if isinstance(geom, Polygon):
        ax.add_patch(
            plt.Polygon(
                list(geom.exterior.coords),
                facecolor=colors.BLUE,
                edgecolor="red",
                hatch="///",
                linewidth=0,
            )
        )
    elif isinstance(geom, MultiPolygon):
        for part in geom.geoms:
            ax.add_patch(
                plt.Polygon(
                    list(part.exterior.coords),
                    facecolor=colors.BLUE,
                    edgecolor="red",
                    hatch="///",
                    linewidth=0,
                )
            )
    else:
        pass


# Add labels
for x, y, label in zip(
    emilia_romagna_provinces.geometry.centroid.x,
    emilia_romagna_provinces.geometry.centroid.y,
    emilia_romagna_provinces["NAME_2"],
    strict=False,
):
    if label == "Reggio Nell'Emilia":
        ax.text(x, y, label, fontsize=10, ha="center")
    else:
        ax.text(x, y, label, fontsize=10)

plt.title("Provinces of Emilia-Romagna, Italy")
plt.axis("off")
plt.savefig("plots/Provinces of Emilia vs Romagna", bbox_inches="tight")
plt.show()
```
